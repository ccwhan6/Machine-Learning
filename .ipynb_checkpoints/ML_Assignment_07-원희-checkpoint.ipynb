```python
import torch.nn as nn
import torch.nn.functional as F
import math
from matplotlib import pyplot

class Linear(nn.Module):

    def __init__(self, num_classes=2):

        super(Linear, self).__init__()

        self.number_class   = num_classes

        _size_image     = 100* 100
        _num1           = 4096
        _num2           = 2048
        _num3           = 512
        _num4           = 64
        
        self.fc1        = nn.Linear(_size_image, _num1, bias=True)
        self.fc2        = nn.Linear(_num1, _num2, bias=True)
        self.fc3        = nn.Linear(_num2, _num3, bias=True)
        self.fc4        = nn.Linear(_num3, _num4, bias=True)
        self.fc5        = nn.Linear(_num4, num_classes, bias=True)

        self.fc_layer1  = nn.Sequential(self.fc1, nn.ReLU(True), nn.Dropout(p=0.4))
        self.fc_layer2  = nn.Sequential(self.fc2, nn.ReLU(True), nn.Dropout(p=0.4))
        self.fc_layer3  = nn.Sequential(self.fc3, nn.ReLU(True), nn.Dropout(p=0.4))
        self.fc_layer4  = nn.Sequential(self.fc4, nn.ReLU(True), nn.Dropout(p=0.4))
        self.fc_layer5  = nn.Sequential(self.fc5, nn.Sigmoid())
        
        # 요 밑에 있는게 아웃풋 레이어
        self.classifier = nn.Sequential(self.fc_layer1, self.fc_layer2, self.fc_layer3, self.fc_layer4, self.fc_layer5)
        
        self._initialize_weight()        
        
    def _initialize_weight(self):
        m = self.fc1
        n = m.in_features
        # 초기값을 assign하는 함수
        m.weight.data.uniform_(- 1.0 / math.sqrt(n), 1.0 / math.sqrt(n))
        if m.bias is not None:
            m.bias.data.zero_()
            
        m = self.fc2
        n = m.in_features
        # 초기값을 assign하는 함수
        m.weight.data.uniform_(- 1.0 / math.sqrt(n), 1.0 / math.sqrt(n))
        if m.bias is not None:
            m.bias.data.zero_()
            
        m = self.fc3
        n = m.in_features
        # 초기값을 assign하는 함수
        m.weight.data.uniform_(- 1.0 / math.sqrt(n), 1.0 / math.sqrt(n))
        if m.bias is not None:
            m.bias.data.zero_()
            
        m = self.fc4
        n = m.in_features
        # 초기값을 assign하는 함수
        m.weight.data.uniform_(- 1.0 / math.sqrt(n), 1.0 / math.sqrt(n))
        if m.bias is not None:
            m.bias.data.zero_()
            
        m = self.fc5
        n = m.in_features
        # 초기값을 assign하는 함수
        m.weight.data.uniform_(- 1.0 / math.sqrt(n), 1.0 / math.sqrt(n))
        if m.bias is not None:
            m.bias.data.zero_()
        '''
        for m in self.modules():
            print(self.modules())
            # weight의 수. 몇개인지
            print(11111111111111)
            print(m)
            n = m.in_features
            # 초기값을 assign하는 함수
            m.weight.data.uniform_(- 1.0 / math.sqrt(n), 1.0 / math.sqrt(n))

            if m.bias is not None:

                m.bias.data.zero_()
        '''

    # forward propagation 정의
    def forward(self, x):

        x = x.view(x.size(0), -1)
        x = self.classifier(x)

        return x
```

### Training and testing

```python
# -----------------------------------------------------------------------------
# import packages
# -----------------------------------------------------------------------------
from __future__ import print_function

import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
import torch.backends.cudnn as cudnn
import argparse
import sys
import os
import numpy as np
import time
import datetime 
import csv
import configparser
import argparse
import platform

import torchvision
from torchvision import datasets, transforms
from torch.autograd import Variable
from random import shuffle

# -----------------------------------------------------------------------------
# load dataset
# -----------------------------------------------------------------------------
transform = transforms.Compose([# transforms.Resize((256,256)),  
                                transforms.Grayscale(),		# the code transforms.Graysclae() is for changing the size [3,100,100] to [1, 100, 100] (notice : [channel, height, width] )
                                transforms.ToTensor(),])
# train_data_path = 'relative path of training data set'
train_data_path = 'horse-or-human/train'
trainset = torchvision.datasets.ImageFolder(root=train_data_path, transform=transform)  

validation_data_path = 'horse-or-human/validation'
valset = torchvision.datasets.ImageFolder(root=validation_data_path, transform=transform)

# change the valuse of batch_size, num_workers for your program
# if shuffle=True, the data reshuffled at every epoch 
loader_train = torch.utils.data.DataLoader(trainset, batch_size=79, shuffle=True, num_workers=2)
loader_test  = torch.utils.data.DataLoader(valset, batch_size=32, shuffle=False, num_workers=2)  

# set_train   = 
# set_test    = 

# -----------------------------------------------------------------------------
# load neural network model
# -----------------------------------------------------------------------------

# from Linear import *
num_classes = 2
model = Linear(num_classes=num_classes)

# -----------------------------------------------------------------------------
# Set the flag for using cuda
# -----------------------------------------------------------------------------

# 쓰실 사람은 쓰시고. 원래 1이었음.
bCuda = 0

if bCuda:
 
    model.cuda()

# -----------------------------------------------------------------------------
# optimization algorithm
# -----------------------------------------------------------------------------

lr = 0.002
# stochastic gradient decsent를 사용하는 optimizer다. 파라미터만 정해주면됨.
optimizer   = optim.SGD(model.parameters(), lr=lr, momentum=0.9)
# loss는 cross entropy다
objective   = nn.CrossEntropyLoss()

# -----------------------------------------------------------------------------
# function for training the model
# -----------------------------------------------------------------------------

def train():

    # print('train the model at given epoch')

    loss_train          = []
    accuracy_train      = []

    model.train()

    correct = 0
    correct_batch = 0
    for idx_batch, (data, target) in enumerate(loader_train):
        
        if bCuda:
        
            data, target    = data.cuda(), target.cuda()

        # 밑에는 x,y 인풋들
        data, target    = Variable(data), Variable(target)

        optimizer.zero_grad()

        output  = model(data)
        loss    = objective(output, target)

        loss.backward()
        optimizer.step()

        loss_train_batch    = loss.item()
        loss_train.append(loss_train_batch)
        pred        = output.data.max(1)[1]
        correct     += pred.eq(target.data.view_as(pred)).cpu().sum()
        correct_batch = pred.eq(target.data.view_as(pred)).cpu().sum()
        accuracy_train_batch = float(correct_batch) / len(data)
        accuracy_train.append(accuracy_train_batch)
        
    loss_train_mean     = np.mean(loss_train)  # mean loss
    loss_train_std      = np.std(loss_train)   # std
    accuracy_train_mean   = float(correct) / len(loader_train.dataset)  # mean accuracy
    accuracy_train_std  = np.std(accuracy_train)
    # print(loss_train_mean)
    # print(accuracy_train_mean)

    return {'loss_train_mean': loss_train_mean, 'loss_train_std': loss_train_std, 
            'accuracy_train_mean': accuracy_train_mean, 'accuracy_train_std': accuracy_train_std}

# -----------------------------------------------------------------------------
# function for testing the model
# -----------------------------------------------------------------------------

def test():

    # print('test the model at given epoch')

    accuracy_test   = 0
    loss_test       = 0
    correct         = 0
    
    model.eval()
    
    for idx_batch, (data, target) in enumerate(loader_test):

        if bCuda:
        
            data, target    = data.cuda(), target.cuda()

        data, target    = Variable(data), Variable(target)

        output  = model(data)
        loss    = objective(output, target)

        loss_test   += loss.item()
        # print(loss_test)
        pred        = output.data.max(1)[1]
        correct     += pred.eq(target.data.view_as(pred)).cpu().sum()
        
    loss_test       = loss_test / 8              # mean loss
    accuracy_test   = float(correct) / len(loader_test.dataset)  # mean accuracy
    # print(loss_test)
    # print(accuracy_test)

    return {'loss_test': loss_test, 'accuracy_test': accuracy_test}

# -----------------------------------------------------------------------------
# iteration for the epoch
# -----------------------------------------------------------------------------
epoch = 15000
breakpoint = 0.0000001
prev_loss  = 0
epoch_list = []
train_loss_list = []
train_loss_std_list = []
train_accuracy_list = []
train_accuracy_std_list = []
valid_loss_list = []
valid_accuracy_list = []

for e in range(epoch):
    
    if e > 700:
        for param_group in optimizer.param_groups:
            param_group['lr'] = 0.001
        
    result_train    = train()
    result_test     = test()

    epoch_list.append(e)
    train_loss_list.append(result_train['loss_train_mean'])
    train_loss_std_list.append(result_train['loss_train_std'])
    train_accuracy_list.append(result_train['accuracy_train_mean'])
    train_accuracy_std_list.append(result_train['accuracy_train_std'])
    valid_loss_list.append(result_test['loss_test'])
    valid_accuracy_list.append(result_test['accuracy_test'])
    
    if epoch > 1 and abs(prev_loss - result_train['loss_train_mean']) < breakpoint:  # breakpoint
        break
    prev_loss = result_train['loss_train_mean']
    
# save stuffs as numpy files
np.save('epoch_list', epoch_list)
np.save('train_loss_list', train_loss_list)
np.save('train_loss_std_list', train_loss_std_list)
np.save('train_accuracy_list', train_accuracy_list)
np.save('train_accuracy_std_list', train_accuracy_std_list)
np.save('valid_loss_list', valid_loss_list)
np.save('valid_accuracy_list', valid_accuracy_list)
    
# lr 설정해줘야되고. epoch이 진행됨에 따라서 적은 lr을 적용한다. sgd에는 하나의 lr만 적용하게끔 하지만 이를 epoch에 따라서 변경할 수 있다.
# mini-batch 그것도 설정해줘야함
# model parameter 초기화도 마음대로

'''
[epoch #####] loss: (training) #####, (testing) #####, accuracy: (training) #####, (testing) #####
epoch마다 그리되
training loss는 iteration마다 구한걸 average하고 iteration마다 구한 loss의 standard deviation도 구해서 투명하게 그린다.
standard deviation함수가 있다.
training acc도 마찬가지로 iteration마다 구한걸 average하고 iteration마다 구한 acc의 standard deviation도 구해서 투명하게 그린다.
'''
```
