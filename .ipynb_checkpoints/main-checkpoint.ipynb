```python
# -----------------------------------------------------------------------------
# import packages
# -----------------------------------------------------------------------------
from __future__ import print_function

import torch
from torch.utils.data import Dataset, DataLoader
import torchvision.transforms as transforms
import torchvision
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
from torch.optim import lr_scheduler
import torch.backends.cudnn as cudnn
import argparse
import sys
import os
import numpy as np
import time
import datetime 
import csv
import configparser
import argparse
import platform

from torchvision import datasets, transforms
from torch.autograd import Variable
from random import shuffle
import matplotlib.pyplot as plt
```

```python
# -----------------------------------------------------------------------------
# load dataset
# -----------------------------------------------------------------------------

set_train   = 'horse-or-human/train'
set_test    = 'horse-or-human/validation'

num_classes = 2

# -----------------------------------------------------------------------------
# load neural network model
# -----------------------------------------------------------------------------

from Linear import *
model = Linear(num_classes=num_classes)

# -----------------------------------------------------------------------------
# Set the flag for using cuda
# -----------------------------------------------------------------------------

bCuda = 0

if bCuda:
 
    model.cuda()
```

```python
# -----------------------------------------------------------------------------
# optimization algorithm
# -----------------------------------------------------------------------------

optimizer   = optim.SGD(model.parameters(), lr=0.01, weight_decay=0.05)
scheduler = lr_scheduler.ReduceLROnPlateau(optimizer, 'min', patience=1, min_lr=2e-4, verbose=1)
objective   = nn.CrossEntropyLoss()

transform = transforms.Compose([transforms.Grayscale(), transforms.ToTensor(),])

#train_data_path = 'relative path of training data set'
trainset = torchvision.datasets.ImageFolder(root=set_train, transform=transform)
# change the valuse of batch_size, num_workers for your program
# if shuffle=True, the data reshuffled at every epoch
loader_train = torch.utils.data.DataLoader(trainset, batch_size=3, shuffle=True, num_workers=1)


valset = torchvision.datasets.ImageFolder(root=set_test, transform=transform)
# change the valuse of batch_size, num_workers for your program
loader_test = torch.utils.data.DataLoader(valset, batch_size=3, shuffle=True, num_workers=1)
```

```python
# -----------------------------------------------------------------------------
# function for training the model
# -----------------------------------------------------------------------------

def train():

    # print('train the model at given epoch')

    loss_train       = []
    accuracy_train   = []
    loss_train_sum   = 0
    correct          = 0
    correct_batch    = 0

    model.train()

    for idx_batch, (data, target) in enumerate(loader_train):

        if bCuda:
        
            data, target    = data.cuda(), target.cuda()

        data, target    = Variable(data), Variable(target)

        optimizer.zero_grad()

        output  = model(data)
        loss    = objective(output, target)

        loss.backward()
        optimizer.step()
        
        #print(loss)

        loss_train_batch    = loss.item() / len(data)
        loss_train.append(loss_train_batch)
        
        pred        = output.data.max(1)[1]
        correct     += pred.eq(target.data.view_as(pred)).cpu().sum()
        correct_batch = pred.eq(target.data.view_as(pred)).cpu().sum()
        accuracy_train_batch = float(correct_batch) / len(data)
        accuracy_train.append(accuracy_train_batch)
        
    loss_train_mean     = np.mean(loss_train)
    loss_train_std      = np.std(loss_train)
    accuracy_train_mean   =  float(correct) / len(loader_train.dataset)
    accuracy_train_std = np.std(accuracy_train)
     
    return {'loss_train_mean': loss_train_mean, 'loss_train_std': loss_train_std, 'accuracy_train_mean' : accuracy_train_mean, 'accuracy_train_std' : accuracy_train_std}
```

```python
# -----------------------------------------------------------------------------
# function for testing the model
# -----------------------------------------------------------------------------

def test():

    # print('test the model at given epoch')

    accuracy_test   = []
    loss_test       = 0
    correct         = 0

    model.eval()

    for idx_batch, (data, target) in enumerate(loader_test):

        if bCuda:
        
            data, target    = data.cuda(), target.cuda()

        data, target    = Variable(data), Variable(target)

        output  = model(data)
        loss    = objective(output, target)

        loss_test   += loss.item()
        pred        = output.data.max(1)[1]
        correct     += pred.eq(target.data.view_as(pred)).cpu().sum()

    loss_test       = loss_test / len(loader_test.dataset)
    accuracy_test   = float(correct) / len(loader_test.dataset)

    return {'loss_test': loss_test, 'accuracy_test': accuracy_test}
```

```python
# -----------------------------------------------------------------------------
# iteration for the epoch
# -----------------------------------------------------------------------------
epoch = 1000

loss_train_mean = []
loss_train_std = []
loss_test = []
accuracy_test = []
accuracy_train_mean = []
accuracy_train_std = []
iteration = []

for e in range(epoch):
        
    result_train    = train()
    result_test     = test()
    
    scheduler.step(result_test['loss_test'], e)
        
    loss_train_mean.append(result_train['loss_train_mean'])
    loss_train_std.append(result_train['loss_train_std'])
    loss_test.append(result_test['loss_test'])
    accuracy_test.append(result_test['accuracy_test'])
    accuracy_train_mean.append(result_train['accuracy_train_mean'])
    accuracy_train_std.append(result_train['accuracy_train_std'])
    iteration.append(e)
    
    print(f"[epoch {e}] loss: (training) {loss_train_mean[e]} , (testing) {loss_test[e]} , accuracy: (training) {accuracy_train_mean[e]} , (testing) {accuracy_test[e]}")
    
    if e >= 1:
        if np.abs(loss_train_mean[e-1] - loss_train_mean[e]) < 0.000001:
            break;
    
```

```python
plt.plot(iteration, loss_train_mean, label = 'training')
plt.plot(iteration, loss_test, label = 'validation')
plt.fill_between(np.arange(1,len(loss_train_mean)+1), np.array(loss_train_mean) - np.array(loss_train_std),np.array(loss_train_mean) + np.array(loss_train_std), alpha = 0.3)
plt.title("Loss")
plt.legend(fontsize='large')
plt.show()
```

```python
plt.plot(iteration, accuracy_train_mean, label = 'training')
plt.plot(iteration, accuracy_test, label = 'validation')
plt.fill_between(np.arange(1,len(accuracy_train_mean)+1), np.array(accuracy_train_mean) - np.array(accuracy_train_std),np.array(accuracy_train_mean) + np.array(accuracy_train_std), alpha = 0.3)
plt.title("Accuracy")
plt.legend(fontsize='large')
plt.show()
```
