```python
import numpy as np
import matplotlib.pyplot as plt
```

## Make Training Set and Testing Set

```python

t1 = np.random.randn(500,2)
t2 = np.random.randn(500,2)
t3 = np.random.randn(500,2)
t4 = np.random.randn(500,2)
std = 3

t1 = std * t1 + 7
t2 = std * t2 + 17
t3 = std * t3 + 7
t4 = std * t4 + 17

trainData = np.zeros((1000,3))
testData = np.zeros((1000,3))

for num in range(1000): #generate training data set with std 3. and centroids are 7 and 17
    if num < 500:
        trainData[num,0] = t1[num,0]
        trainData[num,1] = t1[num,1]
        trainData[num,2] = 1 #label with 1
    else :
        trainData[num,0] = t2[num-500,0]
        trainData[num,1] = t2[num-500,1]  #label with 0
        
for num in range(1000):  #generate testing data set with std 3. and centroids are 7 and 17
    if num < 500:
        testData[num,0] = t3[num,0]
        testData[num,1] = t3[num,1]
        testData[num,2] = 1 #label with 1
    else :
        testData[num,0] = t4[num-500,0]
        testData[num,1] = t4[num-500,1]  #label with 0
```

## Plot data set

```python
to = trainData[:500, :2]
tz = trainData[500:, :2]

plt.title("Training Dataset")
plt.scatter(to[:, 0], to[:, 1], s=10)
plt.scatter(tz[:, 0], tz[:, 1], s=10)
plt.show

to = testData[:500, :2]
tz = testData[500:, :2]

plt.figure()
plt.title("Testing Dataset")
plt.scatter(to[:, 0], to[:, 1], s=10)
plt.scatter(tz[:, 0], tz[:, 1], s=10)
plt.show
```

## define methods for iteration

```python
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

def crossEntropy(l,y):
    return -(l * np.log(y) + ((1 - l) * np.log(1 - y)))      

def loss(l,y):
    return (1 / 1000) * np.sum(crossEntropy(l,y))

def du(z):
    return (1 / 1000) * np.sum((sigmoid(z) - trainData[:,2]) * trainData[: ,0])

def dv(z):
    return (1 / 1000) * np.sum((sigmoid(z) - trainData[:,2]) * trainData[: ,1])

def db(z):
    return (1 / 1000) * np.sum((sigmoid(z) - trainData[:,2]))

def accuracy(l,y):
    cor = 0
    for num in range(1000):
        if (y[num] > 0.5 and l[num] == 1):
            cor = cor + 1
        elif (y[num] <= 0.5 and l[num] == 0):
            cor = cor + 1
    return cor / 1000
```

## make iteration(trainig)

```python
learningRate = 0.015
w = np.array([0.,0.])

b = 0
trainLoss = []
testLoss = []
trainAcc = []
testAcc = []
last = 0
temp = 0
while True:
    trainZ = np.dot(trainData[:, :2],w.T) + b
    testZ = np.dot(testData[:, :2],w.T) + b
    w[0] = w[0] - (learningRate * du(trainZ))
    w[1] = w[1] - (learningRate * dv(trainZ))
    b = b - (learningRate * db(trainZ))
    
    trainLoss.append(loss(trainData[:, 2],sigmoid(trainZ)))
    testLoss.append(loss(testData[:, 2],sigmoid(testZ)))
    
    trainAcc.append(accuracy(trainData[:, 2], sigmoid(trainZ)))
    testAcc.append(accuracy(testData[:, 2], sigmoid(testZ)))
    
    if (abs(trainLoss[temp]-last) < 10e-7):
        break
    else :
        last = trainLoss[temp]
        temp = temp + 1
        continue
```

## plot Loss and Accuracy

```python
plt.plot(trainLoss)
plt.plot(testLoss)
plt.title("Loss")
plt.show()

plt.figure()
plt.plot(trainAcc)
plt.plot(testAcc)
plt.title("Accuracy")
plt.show()
```
